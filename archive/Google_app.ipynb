We have choosen a dataset "Google Play Store Apps", from kaggle. This dataset contains information on apps in the Google Play Store, including variables like the app's rating, the number of downloads, and the category. The Google Play Store Apps dataset from Kaggle contains information on apps in the Google Play Store. It includes 13 variables for each app, including:

1. App: The name of the app.
2. Category: The category of the app (e.g. "Games", "Business").
3. Rating: The rating of the app, on a scale of 1 to 5.
4. Reviews: The number of reviews that the app has received.
5. Size: The size of the app in megabytes (MB).
6. Installs: The number of times that the app has been installed.
7. Type: Whether the app is "Free" or "Paid".
8. Price: The price of the app, if it is a paid app.
9. Content Rating: The age group that the app is intended for (e.g. "Everyone", "Mature 17+").
10. Genres: The genre(s) of the app (e.g. "Action", "Puzzle").
11. Last Updated: The date when the app was last updated.
12. Current Ver: The current version of the app.
13. Android Ver: The minimum Android version required to run the app.

We thought that the following variables could have a linear or polynomial relationship:

1. Rating and number of reviews: It's possible that there is a relationship between the rating of an app and the number of reviews that it has received. Apps with higher ratings may tend to have more reviews, as people are more likely to leave a review if they are satisfied with the app.
2. Size and number of installs: The size of an app may also be related to the number of installs that it has received. Apps that are larger in size may be less appealing to users who have limited storage on their devices, leading to fewer installs.

3. Rating and price: The price of an app may be related to its rating. Users may be more willing to pay for higher quality apps, leading to higher ratings.
   
4. Reviews and number of installs: The number of reviews that an app has received may also be related to the number of installs that it has received. Apps with more reviews may be more popular, leading to more installs.
   
5. Size and rating: The size of an app may be related to its rating. Apps that are larger in size may be more resource-intensive and potentially less smooth to use, leading to lower ratings.
   
6. Genre and number of installs: The genre of an app may also be related to the number of installs that it has received. Apps in certain genres may be more popular than others, leading to more installs.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the Google Play Store Apps dataset
data = pd.read_csv('archive/googleplaystore.csv')

# Set the seaborn style
sns.set()

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2)
"""clean data, remove null values, remove duplicates, remove non-english apps, remove apps that are not free, remove apps that are not in the US"""
data.dropna(inplace=True)

# Linear regression of rating vs. number of reviews
X = data[['Rating']]
y = data['Reviews']
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)
axs[0, 0].scatter(X, y)
axs[0, 0].plot(X, predictions, color='red')
axs[0, 0].set_xlabel('Rating')
axs[0, 0].set_ylabel('Number of Reviews')

# Calculate the mean squared error and R^2 score for the model
mse = mean_squared_error(y, predictions)
r2 = r2_score


def linear_regression(X,Y):
    model = LinearRegression()
    model.fit(X,Y)
    predictions = model.predict(X)
    mse = mean_squared_error(Y, predictions)
    r2 = r2_score(Y, predictions)
    print(f'Mean Squared Error: {mse:.3f}')
    print(f'R^2 Score: {r2:.3f}')
    return mse, r2


```
